# Tokenize

```python
texts_list = df_text_no_dup.text.values.tolist()
texts_list[:5]
```

```python
import re

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

def stopwords_removal(tweet_text, lang):
    
    stop_words = set(stopwords.words(lang))
    words = word_tokenize(tweet_text)
    
    words_filtered = []
    
    for w in words:
        if w not in stop_words:
             words_filtered.append(w)
                
    tweet_text_filtered = ' '; # Separator for join method
    
    return tweet_text_filtered.join(words_filtered)


def clean_tweet(tweet_text):
    
    # Remove https links
    clean_tweet = re.sub(r"https\S+", "", tweet_text)
    # Remove http links
    clean_tweet = re.sub(r"http\S+", "", clean_tweet)
    # Remove pic.twitter links
    clean_tweet = re.sub(r"pic.twitter.com\S+", "", clean_tweet)
    # Remove mentions
    clean_tweet = re.sub(r"@\S+", "", clean_tweet)
    # Replace double spaces with single one
    clean_tweet = clean_tweet.replace('  ', ' ')
    
    clean_tweet = stopwords_removal(clean_tweet, 'italian')
    
    # Replace double spaces with single one
    clean_tweet = clean_tweet.replace('# ', '#')
    
    return clean_tweet
```

```python
texts_cleaned_list = []
for text in texts_list:
    texts_cleaned_list.append(clean_tweet(text))
texts_cleaned_list[:5]
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(texts_cleaned_list)
X_train_counts.shape
```
